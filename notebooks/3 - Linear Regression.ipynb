```python
# IMPORTS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import sklearn
```

```python
# Get the data

from sklearn.datasets import load_diabetes

# This loads a dictionary: dict_keys(['data', 'target', 'DESCR', 'feature_names', 'data_filename', 'target_filename'])
diabetes = load_diabetes()
print(diabetes.DESCR)
```

```python
# Let put the data in a standard pandas dataframe!

# Transform the arrays into an arranged dataframe with the proper column names
df_diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

# We can also set a new Target column
df_diabetes['Target'] = diabetes.target

df_diabetes
```

```python
# Split the data into training and testing set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_diabetes[df_diabetes.columns[:-1]],
                                                    df_diabetes['Target'],
                                                    test_size=0.3)
```

```python
# Train a linear regression with one feature

from sklearn.linear_model import LinearRegression

temp_x_train = X_train['age'].values.reshape(X_train['age'].shape[0], 1)
temp_x_test = X_test['age'].values.reshape(X_test['age'].shape[0], 1)

lr = LinearRegression()
lr.fit(temp_x_train, y_train)

print('Linear Regression Train results: {}'.format(lr.score(temp_x_train, y_train)))
print('Linear Regression Test results: {}'.format(lr.score(temp_x_test, y_test)))
```

```python
# Plot the one variable linear regression

# Plot the data

plt.scatter(temp_x_train, y_train)
plt.plot(temp_x_train, lr.predict(temp_x_train), color='k')
```

```python
# Train a linear regression with multiple variables

lr = LinearRegression()
lr.fit(X_train, y_train)

print('Linear Regression Train results: {}'.format(lr.score(X_train, y_train)))
print('Linear Regression Test results: {}'.format(lr.score(X_test, y_test)))
```

```python
# Plot the data with multiple features

plt.scatter(X_train['age'], y_train)
plt.plot(X_train['age'], lr.predict(X_train), color='k')
```
